### 🎯 Objectives
- Perform data cleaning and preprocessing
- Handle missing values and inconsistencies
- Engineer useful features for analysis

## Importing Required Libraries and Load the Raw Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/telugumovies_dataset_raw.csv')
data.head(5)
```

## Handling Missing Values

```python
data.isnull().sum()
```

```python
data['Year'] = data['Year'].fillna(data['Year'].median())
data['Certificate'] = data['Certificate'].fillna('Unknown')
data['Genre'] = data['Genre'].fillna('Unknown')
data['Overview'] = data['Overview'].fillna('No overview available')
data['Runtime'] = data['Runtime'].replace(0, np.nan)
data['Runtime'] = data['Runtime'].fillna(data['Runtime'].median()).astype(int)
```

```python
data.isnull().sum()
```


Handling Duplicate Values

```python
data.duplicated().sum()
```

## Fixing Data Types and Columns

```python
print(data.dtypes)
```

```python
data['Year'] = data['Year'].astype(int)
data.rename(columns={'No.of.Ratings': 'Num_Ratings'}, inplace=True)
```

```python
print(data.dtypes)
```

## Feature Engineering

```python
#Created a Decade column to analyze trends over time.

data['Decade'] = (data['Year'] // 10) * 10
```

```python
#Calculated movie age based on release year.
data['Movie_Age'] = 2026 - data['Year']
```

```python
#log(1+x) is used because to not make score larger as the num_ratings range from 49 to 99999
data['Popularity_Score'] = data['Rating'] * np.log1p(data['Num_Ratings'])
```

```python
#Extracted the first genre from multi-genre entries to simplify analysis and grouping.

data['Primary_Genre'] = data['Genre'].apply(lambda x: x.split(',')[0].strip())
data.head(5)
```
